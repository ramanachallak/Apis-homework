## Part 1 - Personal Finance Planner

### Collect Crypto Prices Using the `requests` Library

```python
# Set current amount of crypto assets
# YOUR CODE HERE!

my_btc = 1.2
my_eth = 5.3
```

```python
# Crypto API URLs
btc_url = "https://api.alternative.me/v2/ticker/Bitcoin/?convert=CAD"
eth_url = "https://api.alternative.me/v2/ticker/Ethereum/?convert=CAD"
```

```python
# Fetch current BTC price
# YOUR CODE HERE!

btc_data = requests.get(btc_url).json()
btc_price = btc_data["data"]["1"]["quotes"]["USD"]["price"]

# Fetch current ETH price
# YOUR CODE HERE!

eth_data = requests.get(eth_url).json()
eth_price = eth_data["data"]["1027"]["quotes"]["USD"]["price"]

# Compute current value of my crpto
# YOUR CODE HERE!

my_btc_value = my_btc*btc_price
my_eth_value = my_eth*eth_price

# Print current crypto wallet balance
print(f"The current value of your {my_btc} BTC is ${my_btc_value:0.2f}")
print(f"The current value of your {my_eth} ETH is ${my_eth_value:0.2f}")
```

### Collect Investments Data Using Alpaca: `SPY` (stocks) and `AGG` (bonds)

```python
# Set current amount of shares
my_agg = 200
my_spy = 50
```

```python
# Set Alpaca API key and secret
# YOUR CODE HERE!

alpaca_api_key = os.getenv("ALPACA_API_KEY")
alpaca_secret_key = os.getenv("ALPACA_SECRET_KEY")


# Create the Alpaca API object
# YOUR CODE HERE!
alpaca = tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version="v2")
```

```python
# Format current date as ISO format
# YOUR CODE HERE!

#today = pd.Timestamp("2020-07-14", tz="America/New_York").isoformat()

#today_date_ts = pd.Timestamp("today", tz="America/New_York").isoformat() #for today's timestamp

from datetime import datetime
from datetime import timedelta

  
today_date = datetime.now()

end_date = today_date - timedelta(days = 1)

start_date = start_date - timedelta(days=1000)

end_date_iso = pd.Timestamp(end_date, tz="America/New_York").isoformat()
start_date_iso = pd.Timestamp(start_date, tz="America/New_York").isoformat()


today_date_ts = pd.Timestamp("today", tz="America/New_York").isoformat()

start_date_iso

end_date_iso
```

```python
# Set the tickers
tickers = ["AGG", "SPY"]

# Set timeframe to '1D' for Alpaca API
timeframe = "1D"

# Get current closing prices for SPY and AGG
# (use a limit=1000 parameter to call the most recent 1000 days of data) 
# YOUR CODE HERE! I removed the limit = 1000 parameter, since I am setting my Start and End dates 1000 days apart, also while using the get_bars method, the data from the alpaca api was erroneous

df_portfolio = alpaca.get_bars(
    tickers,
    tradeapi.TimeFrame(1, tradeapi.TimeFrameUnit.Day),
    start = start_date_iso,
    end = end_date_iso,
    #limit = 1000
).df

# Preview DataFrame
# YOUR CODE HERE!

df_portfolio
```

```python
# Pick AGG and SPY close prices
# YOUR CODE HERE!

agg_df = df_portfolio[df_portfolio["symbol"] == "AGG"]
spy_df = df_portfolio[df_portfolio["symbol"] == "SPY"]


agg_close_price = agg_df.iloc[-1]["close"]
spy_close_price = spy_df.iloc[-1]["close"]


# Print AGG and SPY close prices
print(f"Current AGG closing price: ${agg_close_price}")
print(f"Current SPY closing price: ${spy_close_price}")
```

```python
# Compute the current value of shares
# YOUR CODE HERE!

my_spy_value = spy_close_price*my_spy
my_agg_value = agg_close_price*my_agg


# Print current value of shares
print(f"The current value of your {my_spy} SPY shares is ${my_spy_value:0.2f}")
print(f"The current value of your {my_agg} AGG shares is ${my_agg_value:0.2f}")
```

### Savings Health Analysis

```python
# Set monthly household income
# YOUR CODE HERE!

monthly_income = 12000

# Consolidate financial assets data
# YOUR CODE HERE!

crypto_assets = my_btc_value + my_eth_value
shares_assets = my_spy_value + my_agg_value


# Create savings DataFrame
# YOUR CODE HERE!

savings_data = {"amount" : [crypto_assets, shares_assets]}

df_savings = pd.DataFrame(savings_data, index = ["crypto", "shares"]) 

# Display savings DataFrame
print(df_savings)
```

```python
# Plot savings pie chart
# YOUR CODE HERE!

df_savings.plot.pie(y = "amount",title="Savings pie chart")
```

```python
# Set ideal emergency fund
emergency_fund = monthly_income * 3

# Calculate total amount of savings
# YOUR CODE HERE!
total_savings = df_savings["amount"]["crypto"] + df_savings["amount"]["shares"]

difference = total_savings - emergency_fund
# Validate saving health
# YOUR CODE HERE!

if(total_savings>emergency_fund):
    print("Congratulations, you have enough money in your fund")
elif(total_savings == emergency_fund):
    print("Congratulations, you have reached your financial goal")
else:
     print(f"You still need to earn ${difference} to reach your financial goal")   

```

## Part 2 - Retirement Planning

### Monte Carlo Simulation

```python
# Set start and end dates of five years back from today.
# Sample results may vary from the solution based on the time frame chosen

#picking the same dates as Part1
start_date_iso

end_date_iso
```

```python
# Get 5 years' worth of historical data for SPY and AGG
# (use a limit=1000 parameter to call the most recent 1000 days of data)
# YOUR CODE HERE! I removed the limit = 1000 parameter, since I am setting my Start and End dates 1000 days apart, also while using the get_bars method, the data from the alpaca api was erroneous 

tickers = ['SPY', 'AGG']

timeframe = "1D"

alpaca_v1 = tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version="v2")


df_stock_data = alpaca_v1.get_bars(
    tickers,
    tradeapi.TimeFrame(1, tradeapi.TimeFrameUnit.Day),    
    start=start_date_iso,
    end=end_date_iso,
    #limit=1000,
).df

df_stock_data
```

```python
# Display sample data
df_stock_data_agg = df_stock_data[df_stock_data["symbol"] == "AGG"]

df_stock_data_spy = df_stock_data[df_stock_data["symbol"] == "SPY"]

df_agg = df_stock_data_agg.drop(['symbol', 'vwap', 'trade_count'], axis = 1)

df_spy = df_stock_data_spy.drop(['symbol', 'vwap', 'trade_count'], axis = 1)


df_agg.sort_index()

df_spy.sort_index()

df_agg.columns = pd.MultiIndex.from_product([['AGG'], df_agg.columns])
df_spy.columns = pd.MultiIndex.from_product([['SPY'], df_spy.columns])

df_agg_spy = pd.concat([df_agg, df_spy], axis = 1)



df_agg_spy


```

```python
# Configuring a Monte Carlo simulation to forecast 30 years cumulative returns
# YOUR CODE HERE!


MC_even_dist = MCSimulation(
    portfolio_data = df_agg_spy,
    weights = [.6,.4],
    num_simulation = 500,
    num_trading_days = 252*30
)

# Print the simulation input data
MC_even_dist.portfolio_data

```

```python
# Printing the simulation input data
# YOUR CODE HERE!
```

```python
# Running a Monte Carlo simulation to forecast 30 years cumulative returns
# YOUR CODE HERE!

MC_even_dist.calc_cumulative_return()
```

```python
# Plot simulation outcomes
# YOUR CODE HERE!

line_plot = MC_even_dist.plot_simulation()
```

```python
# Plot probability distribution and confidence intervals
# YOUR CODE HERE!

dist_plot = MC_even_dist.plot_distribution()
```

### Retirement Analysis

```python
# Fetch summary statistics from the Monte Carlo simulation results
# YOUR CODE HERE!

even_tbl = MC_even_dist.summarize_cumulative_return()

# Print summary statistics
# YOUR CODE HERE!

print(even_tbl)
```

### Calculate the expected portfolio return at the `95%` lower and upper confidence intervals based on a `$20,000` initial investment.

```python
# Set initial investment
initial_investment = 20000

# Use the lower and upper `95%` confidence intervals to calculate the range of the possible outcomes of our $20,000
# YOUR CODE HERE!

ci_lower = round(even_tbl[8]*initial_investment,2)
ci_upper = round(even_tbl[9]*initial_investment,2)


# Print results
print(f"There is a 95% chance that an initial investment of ${initial_investment} in the portfolio"
      f" over the next 30 years will end within in the range of"
      f" ${ci_lower} and ${ci_upper}")
```

### Calculate the expected portfolio return at the `95%` lower and upper confidence intervals based on a `50%` increase in the initial investment.

```python
# Set initial investment
initial_investment_50 = 20000 * 1.5

# Use the lower and upper `95%` confidence intervals to calculate the range of the possible outcomes of our $30,000
# YOUR CODE HERE!

ci_50_lower = round(even_tbl[8]*initial_investment_50,2)
ci_50_upper = round(even_tbl[9]*initial_investment_50,2)

# Print results
print(f"There is a 95% chance that an initial investment of ${initial_investment_50} in the portfolio"
      f" over the next 30 years will end within in the range of"
      f" ${ci_50_lower} and ${ci_50_upper}")
```
